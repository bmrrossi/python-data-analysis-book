```python
#! blockstart
import random
import matplotlib.pyplot as plt
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
#! blockend
```

```python
plt.plot(walk[:100])
```

```python
# Now using numpy
import numpy as np

nsteps = 1000
rng = np.random.default_rng(seed=12345) # gerador de numeros aleatorios
```

```python
draws = rng.integers(0, 2, size=nsteps)
draws
```

```python
steps = np.where(draws == 0, 1, -1)
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
plt.plot(walk[:100])
```
